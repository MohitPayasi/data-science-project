```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
from tensorflow import keras
```

```python
df=pd.read_csv('F:/deep learning/code basic/Churn_Modelling 1.csv')
df
```

```python
df1=df.drop (['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df1.dtypes
```

```python
df1
```

```python
df2= pd.get_dummies(df1['Geography'])
```

```python
df3= pd.concat([df1,df2],axis=1)
```

```python
df3
```

```python
df4=df3.drop(['Geography'],axis=1)
```

```python
df4
```

```python
colu_to_scaled=['CreditScore','Age','Tenure','EstimatedSalary','Balance']
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df4[colu_to_scaled]=scaler.fit_transform(df4[colu_to_scaled])
```

```python
x=df4.drop(['Exited'],axis=1)
y=df4['Exited']
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=1)
```

```python
def ANN (x_train,y_train,x_test,y_test,loss):
    

    model=keras.Sequential([
        keras.layers.Dense(40,input_shape=(12,),activation='relu'),
    
        keras.layers.Dense(40,activation='relu'),
        keras.layers.Dense(1,activation='sigmoid')])
    model.compile(
                 optimizer='adam',
                 loss='binary_crossentropy',
                metrics=['accuracy'])
    model.fit(x_train,y_train,epochs=50)
    
    y_pred=model.predict(x_test)
    y_pred=np.round(y_pred)
    print ("classification report: \n", classification_report(y_test,y_pred))
                                                              
    return y_pred                                                           
```

```python
y_pred=ANN(x_train,y_train,x_test,y_test,'binary_crossentropy')
```

# under sampling

```python
df4['Exited'].value_counts()
```

```python
count_value_0,count_value_1=df4['Exited'].value_counts()
```

```python

df_class_0= df4[df4['Exited']==0]
df_class_1=df4[df4['Exited']==1]
```

```python
df_class_under_0= df_class_0.sample(count_value_1)
```

```python
df_test_under=pd.concat([df_class_under_0,df_class_1],axis=0)
```

```python
df_test_under
```

```python
x=df_test_under.drop(['Exited'],axis='columns')
y=df_test_under['Exited']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=1)
```

```python
y_pred=ANN(x_train,y_train,x_test,y_test,'binary_crossentropy')
```

# over sampling

```python
count_value_0,count_value_1=df5['Exited'].value_counts()
```

```python
df_class_0= df5[df5['Exited']==0]
df_class_1=df5[df5['Exited']==1]
```

```python
x=df_test_under.drop(['Exited'],axis='columns')
y=df_test_under['Exited']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=1)
y_pred=ANN(x_train,y_train,x_test,y_test,'binary_crossentropy')
```
